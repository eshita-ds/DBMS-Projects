### Join `Product` and `Category` Table

```python
product_coll = bikestore_db.products
categ_coll = bikestore_db.category
```

```python
products = product_coll.aggregate(
      [ { '$lookup': { 'from':         'category',
                     'localField':   'Category_id',
                     'foreignField': 'Category_ID',
                     'as':           'category_doc'
                   }
      },
        { '$unwind': '$category_doc' },
     { '$project': { '_id': 0,
                      'Product_Id': 1,
                      'Product_Name': 1,
                      'Model_year': 1,
                      'Price': 1,
                      'Onhand_quantity': 1,
                      'category': '$category_doc.Category_Name'
                    }
      }             
    ]

)

DataFrame(products)
```

### Join the Products, Brands, category collections

```python
product_coll = bikestore_db.products
categ_coll = bikestore_db.category
brand_coll = bikestore_db.brands
```

```python
brands_prod_cat = product_coll.aggregate(
    [ { '$lookup': { 'from':         'brands',
                     'localField':   'Brand_id',
                     'foreignField': 'brand_Id',
                     'as':           'brands_doc'
                   }
      },
      { '$unwind': '$brands_doc' },
      { '$lookup': { 'from':         'category',
                     'localField':   'Category_id',
                     'foreignField': 'Category_ID',
                     'as':           'category_doc'
                   }
      },
      { '$unwind': '$category_doc' },
      { '$project': { '_id': 0,
                      'Product_Id': 1,
                      'Product_Name': 1,
                      'Model_year': 1,
                      'Price': 1,
                      'Onhand_quantity': 1,
                      'Brand': '$brands_doc.brand_name',
                      'category' : '$category_doc.Category_Name'
                    
                    }
      }             
    ]
)
    
DataFrame(brands_prod_cat)
```

#### The category and brand of bike which have product names `Giant Escape 3` and `Trek X-Caliber 9`

```python
Giant_Trek_Cat = product_coll.aggregate(
    [ { '$lookup': { 'from':         'brands',
                     'localField':   'Brand_id',
                     'foreignField': 'brand_Id',
                     'as':           'brands_doc'
                   }
      },
      { '$unwind': '$brands_doc' },
      { '$lookup': { 'from':         'category',
                     'localField':   'Category_id',
                     'foreignField': 'Category_ID',
                     'as':           'category_doc'
                   }
      },
      { '$unwind': '$category_doc' },
      { '$match' : { '$or': [ {'Product_Name': "Giant Escape 3"},
                               {'Product_Name': "Trek X-Caliber 9"}
                             ]
                   }
      },
      { '$project': { '_id': 0,
                      'Product_Id': 1,
                      'Product_Name': 1,
                      'Model_year': 1,
                      'Price': 1,
                      'Brand': '$brands_doc.brand_name',
                      'category' : '$category_doc.Category_Name'
                    
                    }
      }             
    ]
)
    
DataFrame(Giant_Trek_Cat)
```

## Products Brand Cateogory Orders
#### Join the `Product` , `brand`, `category` and `orders` collections.

```python
prod_orders_cat_brand = order_coll.aggregate(
    [ { '$lookup': { 'from':         'products',
                     'localField':   'product_id',
                     'foreignField': 'Product_Id',
                     'as':           'products_doc'
                    }
      },
      { '$unwind': '$products_doc' },
      { '$lookup': { 'from':          'category',
                      'localField':   'products_doc.Category_id',
                      'foreignField': 'Category_ID',
                      'as':           'category_doc'
                    }
      },
      { '$unwind': '$category_doc' },
      { '$lookup': { 'from':         'brands',
                     'localField':   'products_doc.Brand_id',
                     'foreignField': 'brand_Id',
                     'as':           'brands_doc'
                   }
      },
      { '$unwind': '$brands_doc' },
    
      { '$project': { '_id': 0,
                      'order_id':  1,
                      'product_id':  1  ,
                      'Product_Name': '$products_doc.Product_Name',
                      'Category_name' : '$category_doc.Category_Name',
                      'Brand_Name' : '$brands_doc.brand_name'
                    }
      }
   ]
)

DataFrame(prod_orders_cat_brand)
```

### Count the no of orders with Product Id by category Name by Brand Name

```python
prod_orders_group = order_coll.aggregate(
    [ { '$lookup': { 'from':         'products',
                     'localField':   'product_id',
                     'foreignField': 'Product_Id',
                     'as':           'products_doc'
                    }
      },
      { '$unwind': '$products_doc' },
      { '$lookup': { 'from':          'category',
                      'localField':   'products_doc.Category_id',
                      'foreignField': 'Category_ID',
                      'as':           'category_doc'
                    }
      },
      { '$unwind': '$category_doc' },
      { '$lookup': { 'from':         'brands',
                     'localField':   'products_doc.Brand_id',
                     'foreignField': 'brand_Id',
                     'as':           'brands_doc'
                   }
      },
      { '$unwind': '$brands_doc' },
     {
        "$group": {
            "_id": "$product_id",
            "brand_name":{"$first": "$brands_doc.brand_name"},
            "category_name":{"$first": "$category_doc.Category_Name"},
            "count": {"$sum": 1}
        }
        },
      { '$project': { '_id': 0,
                      'product_id': "$_id",
                      'brand_name':  1,
                      'category_name':  1  ,
                      'count': 1
                    }
      }
   ]
)

DataFrame(prod_orders_group)
```
